## (1) Import thư viện, cài đặt cấu hình, lấy và hiển thị dữ liệu
Bước đầu tiên nhằm chuẩn bị môi trường làm việc, bao gồm việc import các thư viện cần thiết cho quá trình xử lý, phân cụm và trực quan hóa dữ liệu, như Pandas, NumPy, Matplotlib, Scikit-learn và Seaborn.


```python
#Nạp thư viện cần thiết
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#Load dữ liệu
wine = pd.read_csv("red_wine.csv")
```

```python
#Xem nhanh dữ liệu
display(wine.head())
print("Kích thước dữ liệu:", wine.shape)
wine.info()
```

#### Nhận xét:
Tập dữ liệu này là một DataFrame sạch, không có giá trị bị thiếu (1599 entries non-null cho tất cả các cột), bao gồm 1599 hàng và 12 cột. Hầu hết các thuộc tính là các phép đo hóa học và vật lý liên tục của rượu vang, được biểu diễn bằng kiểu dữ liệu float64 (11/12 cột). Biến mục tiêu tiềm năng là quality, có kiểu dữ liệu int64, đại diện cho điểm chất lượng của rượu. Do không có giá trị bị thiếu và dữ liệu đã được định dạng tốt, tập dữ liệu này đã sẵn sàng để tiến hành Phân tích Khám phá Dữ liệu (EDA) và xây dựng các mô hình Học máy, chẳng hạn như mô hình hồi quy hoặc phân loại, để dự đoán chất lượng rượu vang dựa trên các thuộc tính hóa học của nó.


## (2) Tiền xử lý dữ liệu
Tiền xử lý dữ liệu (Data Preprocessing) là quá trình làm sạch, chuyển đổi và chuẩn bị dữ liệu thô để tối ưu hóa cho phân tích hoặc xây dựng mô hình Học máy. Nó giúp giải quyết các vấn đề như giá trị bị thiếu, dữ liệu không nhất quán và nhiễu, đảm bảo dữ liệu đầu vào có chất lượng và độ chính xác cao nhất.


### Kiểm tra giá trị trùng lặp
Kiểm tra trùng lặp (Duplicate Checking) là quá trình xác định và loại bỏ các hàng dữ liệu giống hệt nhau trong tập dữ liệu. Mục đích chính là đảm bảo tính chính xác của phân tích và cải thiện hiệu suất xử lý. Trong phân tích dữ liệu, thường sử dụng các công cụ như Pandas để nhanh chóng tìm kiếm và loại bỏ các bản sao.


```python
# Kiểm tra dữ liệu trống
print("Số giá trị bị thiếu trong mỗi cột:\n", wine.isnull().sum())

# Kiểm tra trùng lặp
before = wine.shape[0]
wine = wine.drop_duplicates()
after = wine.shape[0]
print(f"Đã loại {before - after} dòng trùng lặp.")
```

#### Nhận xét:
Tập dữ liệu red_wine_quality chứa 240 bản ghi trùng lặp chính xác. Sự trùng lặp này có thể là do nhiều mẫu rượu vang khác nhau có cùng đặc tính hóa học, hoặc do lỗi nhập liệu. Tuy nhiên, trong tiền xử lý dữ liệu cho Học máy, việc loại bỏ các bản ghi trùng lặp là cần thiết để đảm bảo mỗi tổ hợp thuộc tính chỉ đóng góp một lần vào quá trình huấn luyện mô hình, giúp ngăn ngừa sự thiên lệch và tăng độ tin cậy của kết quả phân tích.


## (3) Thống kê mô tả
Thống kê mô tả (Descriptive Statistics) là quá trình tóm tắt và mô tả các đặc điểm chính của một tập dữ liệu. Nó sử dụng các chỉ số số học như Trung bình (giá trị điển hình), Độ lệch chuẩn (độ phân tán), và các công cụ trực quan để chuyển dữ liệu thô thành thông tin dễ hiểu và có ý nghĩa. Đây là bước đầu tiên để khám phá và hiểu rõ cấu trúc của dữ liệu.


```python
#Thống kê mô tả dữ liệu
desc = wine.describe().T
desc
```

#### Nhận xét: 
-Tổng quan & Kích thước dữ liệu Count (Số lượng): Tất cả 12 thuộc tính đều có 1359 quan sát. Điều này cho thấy sau quá trình tiền xử lý, tập dữ liệu không còn giá trị thiếu (missing values) hay đã được xử lý các hàng có giá trị thiếu/trùng lặp, đảm bảo tính nhất quán cho việc phân tích.
-Nhận xét về Xu hướng Trung tâm (Mean & 50%)
+pH: Giá trị trung bình mean ~ 3.31 và trung vị 50% ~ 3.31 rất gần nhau, cho thấy phân phối của pH khá đối xứng.
+Alcohol: Trung bình mean ~ 10.43 và trung vị 50~ 10.20 cũng gần nhau.
+Citric Acid: Trung bình mean~0.27 lớn hơn trung vị 50% ~0.26 một chút, cho thấy phân phối hơi lệch phải (positive skew).
+Volatile Acidity & Chlorides: Giá trị trung bình lớn hơn đáng kể so với trung vị, báo hiệu phân phối lệch phải mạnh, có thể do sự hiện diện của các giá trị ngoại lai cao.
-Nhận xét về Độ phân tán (Std & Min/Max)
+Total Sulfur Dioxide (Lớn): Có độ lệch chuẩn(std =33.4) và khoảng biến thiên (range: 6.0 → 289.0) rất lớn, cho thấy các giá trị thuộc tính này phân tán rộng và có khả năng chứa nhiều giá trị ngoại lai (outliers) cao.
+Free Sulfur Dioxide (Lớn): Độ lệch chuẩn (std = 10.4) và khoảng biến thiên (1.0→72.0) cũng lớn, xác nhận sự phân tán rộng rãi.
+Density (Nhỏ): Độ lệch chuẩn (std =  0.0018) rất nhỏ, cho thấy các giá trị cực kỳ tập trung xung quanh giá trị trung bình (mean = 0.997), gần như không có sự khác biệt.
